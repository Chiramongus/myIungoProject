```python
import csv
import json

product = {
    "name" : "Harry Potter",
    "price": 3.5,
    "producer" : "Warner Bros",
    "id" : 1
}
product_2 = {
    "name" : "LOTR",
    "price": 45,
    "producer" : "New line cinema",
    "id" : 2
}

products = [product,product_2]

with open("products.csv", "w") as outfile:
    header_names = product.keys()
    writer = csv.DictWriter(outfile, header_names)

    writer.writeheader()
    writer.writerows(products)

```

```python
with open("employees.json", "r") as infile:
    employees = json.load(infile)

with open("employees.csv", "w") as outfile:
    headers = employees[0].keys()
    write = csv.DictWriter(outfile, headers)

    write.writeheader()

    departments = ['Marketing',
    'Engineering - Front End',
    'Engineering - Back End',
    'Customer Research',
    'Governance',
    'Distribution']

    def get_dept():
        user_input = input("Please enter the department you want to write to the CSV: ")
        if user_input in departments:
            for i in range(len(employees)):
                if employees[i]["department"] == user_input:
                    write.writerow(employees[i])
        else:
            get_dept()
    get_dept()       





```

```python
import requests
from bs4 import BeautifulSoup

url = "https://webscraper.io/test-sites/e-commerce/allinone"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html")

```

```python
print(soup.find("p", {"class" : "description card-text"}))
print(soup.find_all("h2"))
print(soup.find_all("a", {"class" : "title"}))

items = [{
    "name" : soup.find("a", {"class" : "title"}).text,
    "description" : soup.find("p", {"class" : "description card-text"}).text,
    "price" : soup.find("h4", {"class" : "price float-end card-title pull-right"}).text
}]
 



namelist = soup.find_all("a", {"class" : "title"})

description_list = soup.find_all("p", {"class" : "description card-text"})

price_list = soup.find_all("h4", {"class" : "price float-end card-title pull-right"})


for i in range(len(namelist)):
    name = namelist[i].text
    description = description_list[i].text
    price = price_list[i].text

    items.append({
        "name" : name,
        "description" : description,
        "price" : price
    })



with open("scraped-info.json", "w") as outfile:
    json.dump(items, outfile, indent=4)
```
